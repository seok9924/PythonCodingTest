```python
import copy


a= input_stone1
b= dogs1

copystone=copy.deepcopy(a)

for i in b:
    name=i['이름'] #이름
    jump=int(i['점프력']) # 점프력
    weight=int(i['몸무게']) # 무게
    length=len(copystone) #4
    jump_count=length//jump #점프길이 / 점프력
    k =0 
    while(k<jump_count):
        k+=1
        onejump=(k*jump-1)
        copystone[onejump]-=weight
    if min(copystone)>=0:
        print(name)
        print(copystone)
    else:
        copystone=copy.deepcopy(a)


```
